```python
import pandas as pd

print("1. Loading Evaluation Results and Ground Truth...")
results_path = "../data/benchmark/full_evaluation_results.csv"
benchmark_path = "../data/benchmark/physics_evaluation_benchmark.csv"

try:
    results_df = pd.read_csv(results_path)
    benchmark_df = pd.read_csv(benchmark_path)
    
    merged_df = pd.merge(results_df, benchmark_df[['question', 'answer']], 
                         left_on='Question', right_on='question', how='left')
    
    print("2. Calculating Strict Exact-Match Metrics...")
    correct_count = 0
    total_questions = len(merged_df)
    
    print("="*50)
    print("TRUE FINAL EVALUATION SCORING")
    print("="*50)
    
    for index, row in merged_df.iterrows():
        question = row['Question']
        # Clean both to ensure no stray spaces
        prediction = str(row['LLM_Prediction']).strip()
        ground_truth = str(row['answer']).strip()
        
        # EXACT MATCH RULE: "1" must equal "1"
        is_correct = (prediction == ground_truth)
        
        if is_correct:
            correct_count += 1
            status = "PASSED"
        else:
            status = "FAILED"
            
        print(f"Q{index+1}: {status}")
        print(f"  Expected : {ground_truth}")
        print(f"  Predicted: {prediction}")
        print("-" * 30)
        
    accuracy = (correct_count / total_questions) * 100
    print(f"\nTRUE TOTAL ACCURACY: {accuracy:.2f}% ({correct_count}/{total_questions})")
    print("="*50)

except Exception as e:
    print(f"Error: {e}")
```
