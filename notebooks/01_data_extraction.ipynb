```python
import fitz  # PyMuPDF
import os

# 1. Define the location of your PDF file
# The "../" means we go up one folder from 'notebooks' and then into 'data/raw'
pdf_path = "../data/raw/university-physics-volume-1.pdf"

def check_page_extraction(file_path, page_num):
    print(f"Attempting to read {file_path}...")
    try:
        # Open the PDF document
        doc = fitz.open(file_path)
        
        # Load the specific page (e.g., page 150)
        page = doc.load_page(page_num)
        
        # Extract text from the page
        text = page.get_text()
        
        print(f"\n--- SUCCESSFULLY READ PAGE {page_num} ---")
        print("Displaying the first 1000 characters:\n")
        print(text[:1000])
        
    except Exception as e:
        print(f"Oops, an error occurred: {e}")

# 2. Let's test it on page 150 (usually contains core material and formulas)
check_page_extraction(pdf_path, 150)
```
